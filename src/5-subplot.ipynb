### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
pie_values = [30, 20, 25, 25]

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line chart
axes[0, 0].plot(x, y, color='blue', linestyle='-', linewidth=2)
axes[0, 0].set_title("Line Chart")

# Bar chart
axes[0, 1].bar(categories, values, color='green')
axes[0, 1].set_title("Bar Chart")

# Scatter plot
axes[1, 0].scatter(x_scatter, y_scatter, color='red')
axes[1, 0].set_title("Scatter Plot")

# Pie chart
axes[1, 1].pie(pie_values, labels=categories, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'purple'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()
```
